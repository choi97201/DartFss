```python
from selenium import webdriver
from pykrx import stock
import pandas as pd
import time
from bs4 import BeautifulSoup
import cjw_maria

host = "localhost"
user = "root"
password = "1234" 
db = "jamoo"

maria = cjw_maria.MariaDB(host, user, password, db)
dgData = maria.showData('select * from dg2017')

stock_amount_2016 = stock.get_market_cap_by_ticker("20161229 ")[['상장주식수']]
code_list = list(stock_amount_2016.index)

driver = webdriver.Chrome('chromedriver.exe')

df = []
dg2017 = []
dg2018 = []
dg2019 = []
dg2020 = []
error_code = []
for code in code_list[1000:]:
    if code in list(dgData['종목코드']):
        continue
    print(code)
    try:
        url = 'https://signalm.sedaily.com/StockFS/{}/goingconcern'.format(code)
        driver.get(url)
        time.sleep(3)
        html = driver.page_source
        soup = BeautifulSoup(html,'html.parser')
        div = soup.select_one("#StockInfoDetailTab_2 > div.sec.sec2 > div > div > div.div_scroll > table")
        table = pd.read_html(str(div))[0]
    except Exception as e:
        error_code.append(code)
    else:
        tmp = []
        for c in table.columns:
            if '2017' in c or '2018' in c or '2019' in c or '2020' in c:
                tmp.append(c)
        table = table[['항목'] + tmp]

        table.columns = [c.replace('.12(연결)','') for c in table.columns]
        table.columns = [c.replace('.12(별도)','') for c in table.columns]

        t1 = table[table['항목'] == '단기차입금']
        t2 = table[table['항목'] == '장기차입금']

        t = pd.concat([t1, t2])
        t = t.transpose().reset_index(drop=False)
        t.columns = t.iloc[0]
        t = t.rename(columns={'항목': '년도'})
        t = t.drop(0).reset_index(drop=True)
        try:
            tmp = list(t[t['년도'] == '2017'][['단기차입금', '장기차입금']].reset_index(drop=True).iloc[0])
            maria.insertData('dg2017', tuple([code] + list(map(int, tmp))))
            maria.commitDB()
            print(tuple([code] + list(map(int, tmp))))
        except Exception as e:
            maria.insertData('dg2017', tuple([code] + [-9999999, -9999999]))
            maria.commitDB()
            print(e.args)
            pass
        try:
            tmp = list(t[t['년도'] == '2018'][['단기차입금', '장기차입금']].reset_index(drop=True).iloc[0])
            maria.insertData('dg2018', tuple([code] + list(map(int, tmp))))
            maria.commitDB()
        except Exception as e:
            maria.insertData('dg2018', tuple([code] + [-9999999, -9999999]))
            maria.commitDB()
            print(e.args)
            pass
        try:
            tmp = list(t[t['년도'] == '2019'][['단기차입금', '장기차입금']].reset_index(drop=True).iloc[0])
            maria.insertData('dg2019', tuple([code] + list(map(int, tmp))))
            maria.commitDB()
        except Exception as e:
            maria.insertData('dg2019', tuple([code] + [-9999999, -9999999]))
            maria.commitDB()
            print(e.args)
            pass
        try:
            tmp = list(t[t['년도'] == '2020'][['단기차입금', '장기차입금']].reset_index(drop=True).iloc[0])
            maria.insertData('dg2020', tuple([code] + list(map(int, tmp))))
            maria.commitDB()
        except Exception as e:
            maria.insertData('dg2020', tuple([code] + [-9999999, -9999999]))
            maria.commitDB()
            print(e.args)
            pass
    

```
