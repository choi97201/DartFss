```python
print('전체: {}'.format(orgin_df.shape[0]))

a_df.to_excel('기본조건_{}.xlsx'.format(today), index=False, encoding='utf-8-sig')
print('기본조건: {}'.format(a_df.shape[0]))

comb_ac = pd.merge(a_df[['종목코드', '최신가격', '주당가치분석']], c_df, on=['종목코드', '주당가치분석'])
comb_ac.to_excel('영업이익(억)_{}.xlsx'.format(today), index=False, encoding='utf-8-sig')
print('영업이익: {}'.format(comb_ac.shape[0]))

comb_ad = pd.merge(a_df[['종목코드', '최신가격', '주당가치분석']], d_df, on=['종목코드', '주당가치분석'])
comb_ad.to_excel('부채비율{}(억)_{}.xlsx'.format(dept_rate, today), index=False, encoding='utf-8-sig')
print('부채비율{}: {}'.format(dept_rate, comb_ad.shape[0]))

comb_acd = pd.merge(comb_ac[['종목코드', '최신가격', '주당가치분석', '2018영업이익성장률', '2019영업이익성장률']], comb_ad, on=['종목코드', '주당가치분석', '최신가격'])
comb_acd.to_excel('영업이익_부채비율{}(억)_{}.xlsx'.format(dept_rate, today), index=False, encoding='utf-8-sig')
print('영업이익_부채비율{}: {}'.format(dept_rate, comb_acd.shape[0]))
```

```python
from selenium import webdriver
import time
from bs4 import BeautifulSoup
import pandas as pd
import collector
import numpy as np

main = collector.Collector('fnguide')
main._setMaria()

code = '105560'
webdriver_options = webdriver.ChromeOptions()
webdriver_options.add_argument('headless')
chromedriver = 'chromedriver.exe'
driver = webdriver.Chrome(chromedriver, options=webdriver_options)

main.maria.createTable('roe_202103', ['종목코드', 'ROE', '날짜'], ['VARCHAR(20)', 'FLOAT', 'VARCHAR(20)'])
done_list = main.getDoneList('roe_202103')
for code in main.code_list:
    if code in done_list:
        continue
    try:
        url = 'https://comp.fnguide.com/SVO2/ASP/SVD_FinanceRatio.asp?pGB=1&gicode=A{}&cID=&MenuYn=Y&ReportGB=&NewMenuID=104&stkGb=701'.format(code)
        driver.get(url)
        html = driver.page_source
        df_list = pd.read_html(html)
        df = df_list[0]
        data = tuple([code, df[df[df.columns[0]].str.contains('ROE')][df.columns[-1]].iloc[0], str(df.columns[-1])])

        main.maria.insertData('roe_202103', data)
        main.maria.commitDB()
    except:
        pass
```
