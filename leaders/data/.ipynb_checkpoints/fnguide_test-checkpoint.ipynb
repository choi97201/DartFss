```python
from cjw_mysql import *
from pykrx import stock
from utils import *
import requests
from tqdm import tqdm
import time
from fnguide import *
```

```python
data = {}
fn = Data()

data_path = 'C:/Users/choi97201/Downloads/제조업'
file_list = os.listdir(data_path)
file_list = [curr_file for curr_file in file_list if '.csv' in curr_file]

for curr_file in file_list:
    
    num = 7
    if curr_file == '부채비율.csv':
        num = 8
    
    subject = curr_file.replace('.csv', '')
    df = pd.read_csv(os.path.join(data_path, curr_file))
    df.columns = df.iloc[num]
    df = df.drop(num).reset_index(drop=True)
    df = df.iloc[num:].reset_index(drop=True)
    cols = list(df.columns)
    df = df[[cols[0]] + cols[4:]]
    data[subject] = df
```

```python
for wish in list(data.keys()):
    display(data[wish])
```

```python
for wish in list(data.keys()):
    data[wish].to_csv(f'C:/data/subject_{wish}.csv', index=False, encoding='cp949')
```

```python
db = Maria()
db.setMaria(host='localhost', user='root', password='sa1234', db='ohlcv', charset='utf8')
stock_info = db.mariaShowData('stock_info')
new_stock_info = stock_info.values.tolist()
driver = setChrome(False)
```

```python
for curr_list in tqdm(new_stock_info):
    curr_sector = 'None'
    curr_code = curr_list[0]
    curr_url = f'https://finance.naver.com/item/coinfo.nhn?code={curr_code}'
    driver.get(curr_url)
    try:
        driver.switch_to.frame('coinfo_cp')
        elements_list = driver.find_elements_by_css_selector('.line-left')
        for curr_element in elements_list:
            if 'WICS' in curr_element.text:
                curr_sector = curr_element.text.replace('WICS : ', '')
                break
    except:
        pass
    curr_list.append(curr_sector)
```
